#การทำให้ข้อมูลเป็นมาตรฐาน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'วิศวกรรม', 'วิศวฯ', 'วิศวกรรมศาสตร์','Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

```python
def standardize_faculty(name):
    if name in ['ชาย', 'ช', 'M', 'Male']:
        return 'M'
    elif name in ['หญิง', 'ญ', 'F','Female']:
        return 'F'
    else:
        return name
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_faculty)
```

```python
df
```

#การจัดการค่าผิดปกติ

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df= df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df
```

```python
df = df[(df['GPA'] >= 0.00) & (df['GPA'] <= 4.00)]
```

```python
df
```

#การรวม Features

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

df[['ชื่อ', 'BMI', 'สุขภาพ']]
```
